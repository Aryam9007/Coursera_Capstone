```python
import pandas as pd, numpy as np, requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
source = requests.get(url).text
soup = bs(source, 'html')
table = soup.find('table', class_='wikitable sortable')
headers = [header.text.replace('\n','') for header in table.find_all('th')]
data = []
for row in table.find_all('tr'):
    col = row.find_all('td')
    col = [e.text.replace('\n','') for e in col]
    row = [e if e !='Not assigned' else np.nan for e in col if e]
    if len(row)>0:
            data.append(row)
df = pd.DataFrame(data, columns=headers)
```

## Let us generate a map of Toronto Using our data 

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df_geo['Latitude'],df_geo['Longitude'],df_geo['Borough'],df_geo['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='red',
    fill=True,
    fill_color='#cc3131',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Now Let us cluster the data points

```python
k=5
df_km = df_geo # creating a backup of data
toronto_kmclustering = df_km.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_kmclustering)
kmeans.labels_
df_km.insert(0, 'Cluster Labels', kmeans.labels_)
df_km
```

```python
df_km
```

```python
# creating the map
map_km = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# setting color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# adding markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_km['Latitude'], df_km['Longitude'], df_km['Neighborhood'], df_km['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_km)
       
map_km

```
